# 시도/ 시군구 로 주소 나눠주기

## 시도 / 시군구 정보를 나누고 싶을 때
.item() : (index, value) 형태의 Tuple 을 만든다

### 시도 칼럼 만들어주기

```python
def get_sido(x):
    return x.split()[0]
```

```python
list(train['region'].items())
```

```python
sido = []
for i, v in train['region'].items():
    sido.append(get_sido(v))
```

```python
sido
```

```python
train['sido'] = sido
```

### 시군구 칼럼 만들어주기

```python
def get_sigungu(x):
    return x.split()[1]
```

```python
# 함수 확인
sample = train['region'][10]

get_sigungu(sample)
```

```python
sigungu = []
for i, v in train['region'].items():
    sigungu.append(get_sigungu(v))
```

# for loop 관련 Tip

```python
# break: for loop 테스트하기 
sigungu = []
for i, v in train['region'].items():
    sigungu.append(get_sigungu(v))
    break
```

```python
# 리스트로 넣어서 불러올 수도 있지만, 
# 그러면 리스트로 미리 만드는 시간과 저장하는 메모리가 소요된다
# cf. .items() 를 쓰면 해당 아이템을 매번 하나씩 만들어서 가져온다. 
sigungu = []
for v in list(train['region']):
    sigungu.append(get_sigungu(v))
```

#### 계속

```python
sigungu
```

```python
train['sigungu'] = sigungu
```

```python
train.sample()
```

```python
train[train['sido'] == '제주']
```

## apply()를 사용하면 for loop 없이
apply 보다는 groupby 등 행렬 연산이 훨씬 빠르다

```python
train['sido'] = train['region'].apply(get_sido)
```

```python
train['sigungu'] = train['region'].apply(get_sigungu)
```

# lambda를 사용하면 함수 def 만들 필요 없이

#### 시도

```python
get_sido = lambda x: x.split()[0]
```

```python
train['sido'] = train['region'].apply(get_sido)
```

#### 시군구

```python
get_sigungu = lambda x: x.split()[1]
```

```python
train['sigungu'] = train['region'].apply(get_sigungu)
```

### 한줄로 lambda

```python
train['sido'] = train['region'].apply(lambda x: x.split()[0])
```

```python
train['sigungu'] = train['region'].apply(lambda x: x.split()[1])
```

# 응용

```python
train['transacted_date'] = pd.to_datetime(train['transacted_date'])
```

### Q1. 컬럼 `transacted_date`를 참조하여 결제가 있었던 연도를 뽑아내 `year`라는 칼럼에 저장하세요.

```python
train['year'] = train['transacted_date'].dt.year
train[['transacted_date', 'year']].sample()
```

### Q2. 컬럼 `transacted_date`를 참조하여 결제가 있었던 월을 뽑아내 `month`라는 칼럼에 저장하세요.

```python
train['month'] = train['transacted_date'].dt.month
train[['transacted_date', 'month']].sample()
```
