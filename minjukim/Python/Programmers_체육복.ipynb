````python
## 통과
def solution(n, lost, reserve):
    
    # 순서대로
    lost.sort()
    reserve.sort()

    # 잃어버렸지만 여벌 있는 경우
    still_got_clothes = set(lost) & set(reserve)
    lost = set(lost) - still_got_clothes
    reserve = set(reserve) - still_got_clothes
    
    # 진짜 잃어버린애, 진짜 여벌 있는 애 카운트
    pure_lost_count = len(lost)
    pure_reserve_count = len(reserve)
    
    # # 앞 뒤 번호에 Reserve 가 있는지, 빌리면 Remove
    for number in lost:

        if (number - 1) in reserve:
            reserve.remove(number - 1)
            
        elif (number + 1) in reserve:
            reserve.remove(number + 1)
            
    answer = n - pure_lost_count + pure_reserve_count - len(reserve)
    
    return answer
````

-|실행
-|-
sorted(배열)| 배열 을 그대로 두고 정렬된 것을 복제
배열.sort()| 배열 을 복제

cf. tuple 에는 적용 안된다. 값 변경이 안되니까. reverse 등등도 마찬가지 

````python
def solution(n, lost, reserve):

    lost = [i for i in lost if i not in reserve]
    reserve = [i for i in reserve if i not in lost]
    
    answer = n - len(lost)
    
    lost.sort()
    reserve.sort()

    for i in lost:
        for j in reserve:
            if i - 1 == j or i + 1 == j:
                answer +=1
                reserve.remove(j)
                break
            
    return answer
````

````python
# 개쩌는 코드
def solution(n, lost, reserve):
    _reserve = [r for r in reserve if r not in lost]
    _lost = [l for l in lost if l not in reserve]
    for r in _reserve:
        f = r - 1
        b = r + 1
        if f in _lost:
            _lost.remove(f)
        elif b in _lost:
            _lost.remove(b)
    return n - len(_lost)
````
